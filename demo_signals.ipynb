# Visualizing audio signals
### Import the audio and plot it

```python
fs, x = read('handel.wav')
print("Sampling frequency = ", fs)
print("Signal size = ", x.shape)
```

```python
# Plot audio 
plt.plot(x)
plt.title('Plot of audio amplitude over time')
plt.xlabel('samples')
plt.show()
```

In the following code block, you can manipulate the start index and the length of the window to zoom in on different parts of the signal.

```python
start_index = 0 # Change this!
window_length = 20 # Change this!

print('x = ', x[start_index:start_index+window_length])
plt.stem(x[start_index:start_index+window_length])
plt.title('Plot of audio amplitude over time, zoomed in')
plt.xlabel('samples')
plt.show()
```

# Unit impulse and shifted unit impulse

$ x_0 = \delta[n] $

$ x_1 = \delta[n-m] $

The unit impulse function $\delta[n]$ has a value of $1$ at $n = 0$ and a value of $0$ everywhere else. We can shift the unit impulse by introducing a new value, $m$.

The following code block plots the unit impulse within the range of $-5 \leq n \leq 5$. Thus you can choose values of $m$ between $-5$ and $5$ to see how the impulse is shifted.

```python
n = np.linspace(-5,5,11)
m = 2 # Change this!

x0 = np.zeros_like(n)
x1 = np.zeros_like(n)
x0[5] = 1
x1[5+m] = 1

plt.subplot(211)
plt.stem(n,x0)
plt.title('x1')
plt.subplot(212)
plt.stem(n,x1)
plt.title('x2')
plt.show()
```

# Sounding out middle C
Middle C (or C4) has a frequency of 261.6 Hz.

$ x = \sin(2\pi\cdot261.6\cdot n)$

Here we set a variable $f_1$ to the desired frequency. You can change the value to sound out different notes. For example, Middle A is 440 Hz.

```python
fs = 4000  # sampling frequence
T = 5      # signal length in seconds
# Discrete sampling locations between [0, T] seconds at fs sampling frequency
n = np.linspace(0, T, T*fs,endpoint=False)  
f1 = 261.6 # Change this number for different notes
x = np.sin(2*np.pi*f1*n)
plt.plot(n[:100],x[:100])
plt.xlabel('seconds')
plt.show()
```

```python
Audio(x, rate = fs)
```
